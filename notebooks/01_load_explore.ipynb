```python
# Importing the required libraries for process mining and data handling
from pm4py.objects.log.importer.xes import importer as xes_importer
from pm4py.objects.conversion.log import converter as log_converter
from pathlib import Path
import pandas as pd

# Defining the relative path to the raw event log

xes_path = Path("../data/raw/BPI_Challenge_2017.xes")

# Loading the event log using pm4py
# The imported log object contains all cases, events, and attributes
log = xes_importer.apply(str(xes_path))

# Converting the event log to a pandas DataFrame
# This allows for easier inspection, filtering, and metric calculation
df = log_converter.apply(log, variant=log_converter.Variants.TO_DATA_FRAME)

# Displaying basic information about the dataset
print("✅ Event log successfully loaded!")
print(f"Rows: {len(df):,} | Columns: {len(df.columns)}")

# Showing the first few rows for an initial overview
df.head()
```
